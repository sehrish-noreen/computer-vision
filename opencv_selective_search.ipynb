```python
import cv2
import selectivesearch
import matplotlib.pyplot as plt
```

## Example 1

```python
image_path = "fruits-data/images/gettyimages-1840650286-640x640.jpg"
image = cv2.imread(image_path)
# convert to rgb
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# bounding boxes
boxes = [
    (100, 50, 200, 150),
    (200, 150, 300, 250)
]
# drawing boxes
for (x, y, w, h) in boxes:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 247, 0), 5)
plt.imshow(image)
```

```python
image = cv2.imread(image_path)
print(image.shape) # height, width, channels
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# resize image to reduce computation time
new_height = int(image.shape[1] / 4)
new_width = int(image.shape[0] / 2)
image = cv2.resize(image, (new_width, new_height))
plt.imshow(image)
```

```python
print(image.shape)
```

```python
# selective search
image_label, regions = selectivesearch.selective_search(image, scale=500, sigma=0.9, min_size=50)
# copy of image for visualization
img_region_proposals = image.copy()
# draw boxes on proposed regions
for region in regions:
    x, y, w, h = region['rect']
    cv2.rectangle(img_region_proposals, (x,y), (x+w, y+h), (255, 247, 0), 1)
plt.imshow(img_region_proposals)
```

## Example 2

```python
image_path = "images/cat.jpg"
image = cv2.imread(image_path)
# convert to rgb
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image_labels, regions = selectivesearch.selective_search(image, scale=700, sigma=0.9, min_size=500)
img_region_proposals = image.copy()
for region in regions:
    x, w, y, h = region['rect']
    cv2.rectangle(img_region_proposals, (x, y), (x+w, y+h), (255, 247, 0), 1)
plt.imshow(img_region_proposals)
```
